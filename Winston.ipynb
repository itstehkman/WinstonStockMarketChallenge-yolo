```python
import pandas as pd
import numpy as np
from ggplot import *
pd.options.display.max_columns = 999
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
features = data[list(data.columns[1:26])]
```

```python
features
```

```python
print '%d rows with nan'%sum(features['Feature_1'].isnull())
ggplot(aes(x='Feature_1'), data = features) + geom_histogram()
```

```python
complete_features = features.dropna()
```

```python
complete_features.corr()
```

```python
#Visualize Correlation Matrix
corr = complete_features.corr()
import matplotlib.pyplot as plt

#Get the subplots
fig, ax = plt.subplots(figsize=(8,8))

#draw
ax.matshow(corr)
fig.colorbar(ax.matshow(corr))

#Customize
plt.xticks(np.arange(0, 24, 2))
plt.yticks(np.arange(0, 25, 2))
plt.title('Correlation Matrix')
```

So as you can see, the red and dark blue cell pairs are highly correlated and we hsould probably drop them to clean up our features and not clutter the regression.

Let's try to filter to get just the least correlated features, which seems to be most green cell pairs and light blue pairs.

```python
maxCorr = .4
important_features = corr[corr < maxCorr][corr > -maxCorr]
important_features

```

```python
#pick out features with mostly non NaN in them
important_features = important_features.dropna(thresh=21)
features[important_features.index]

```
