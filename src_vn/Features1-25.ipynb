```python
features = data[list(data.columns[1:26])]
```

# Which features are most complete? 

```python
feature_completeness = pd.Series()
for feature in features.columns:
    feature_completeness[feature] = 100.0 * len(features[feature].dropna())/len(features[feature])
```

Below: What percent of examples have a valid feature field?:

```python
feature_completeness
```

##  Let's drop any features with less than 95% valid examples:

```python
compact_feature_list = list(feature_completeness[feature_completeness>95].index)
compact_features = features[compact_feature_list].dropna()
print 'Out of %d examples, %d of them have a complete set of data for our compact features.'%(len(features),len(compact_features))
```

```python
feature_correlations = compact_features.corr()
feature_correlations
```

## Find highly correlated features: 

The below pairs are possibly redundant:

```python
for feature in feature_correlations.columns:
    single_feature = feature_correlations[feature].abs()
    high_corr = single_feature[(single_feature>.8) & (single_feature != 1)]
    for high_corr_feature in high_corr.index:
        print (feature,high_corr_feature,high_corr[high_corr_feature])
    
```
