```python
quarter_day = data[['Id'] + list(data.columns[28:147])]
quarter_day.index = quarter_day.Id
quarter_day.drop('Id', axis=1, inplace = True)
quarter_day = quarter_day + 1
#quarter_return = quarter_day.multiply(axis = 0)
quarter_day.fillna(1, inplace = True)
quarter_day['quarter_return'] = quarter_day.apply(lambda x: np.array(x).prod(), axis = 1)
quarter_day['Id'] = quarter_day.index

quarter_day.head()
```

```python
filtered_data = pd.merge(filtered_data, quarter_day[['Id','quarter_return']], on = 'Id')
filtered_data.head()
```

```python
filtered_data.to_csv('First_Dataset.csv')
```
