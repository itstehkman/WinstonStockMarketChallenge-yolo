# Let's look at returns: 

```python
import pandas as pd
import numpy as np
import os
from ggplot import *
from __future__ import division
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
%matplotlib inline
```

```python
data_path = 'C:/Users/vnguyen/Dropbox/Winton/data_vn'
file_name = 'train.csv'
data = pd.read_csv(os.path.join(data_path,file_name))
```

```python
data.head()
```

```python
for col in data.columns: print col
```

```python
returns_list = [col for col in data.columns if 'Ret' in col]
returns_list
```

```python
returns = data[returns_list]
```

```python
returns_completeness = pd.Series()
for ret in returns.columns:
    returns_completeness[ret] = 100.0 * len(returns[ret].dropna())/len(returns[ret])
```

```python
returns_completeness.describe()
```

```python
minute_returns = returns.drop(['Ret_MinusTwo','Ret_MinusOne','Ret_PlusOne','Ret_PlusTwo'],axis = 1)
```

```python
minute_returns.head()
```

```python
plot_data = pd.melt(minute_returns)
plot_data.head()
```

```python
plot_data.describe()
```

```python
ggplot(aes(x='value'), data = plot_data[np.abs(plot_data.value)<.008])\
+ geom_histogram(binwidth = .005)\
+ ggtitle('Minute Returns')
```

```python
plot_data = pd.melt(data[['Ret_MinusTwo','Ret_MinusOne','Ret_PlusOne','Ret_PlusTwo']])
plot_data.head()
```

```python
plot_data.describe()
```

```python
plot_data.groupby('variable').agg([np.mean,np.std,lambda x: np.mean(np.abs(x))])
```

```python
ggplot(aes(x='value'), data = plot_data[np.abs(plot_data.value)<.2])\
+ geom_histogram(binwidth = .005)\
+ ggtitle('Daily Returns')
```
